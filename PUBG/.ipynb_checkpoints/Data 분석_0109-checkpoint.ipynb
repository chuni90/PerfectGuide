* user_df의 각 feature에 대한 설명
<hr>
* DBNOs : 플레이어가 기절한 횟수 (solo면 0)
* assists : 플레이어가 적에게 데미지를 입혔고, 해당 적이 팀원에게 죽은 수
* boosts : 부스트 아이템을 사용한 횟수(진통제, 에너지드링크, 주사기)
* kills : 플레이어가 적을 죽인 횟수
* killStreaks : 연속 킬 수
* headshotKills : 헤드샷으로 죽인 플레이어 수
* roadKills : 운전으로 죽인 킬 수
* teamKilss : 팀킬 수
* longestKill : 가장 멀리서 적을 죽인 거리
* damageDealt : 적에게 가한 총 데미지(자해 데미지는 차감됨)
* heals : 힐링아이템 사용 횟수(의료용키트, 구급상자,붕대)
* revives : 이 플레이어가 팀원을 되살린 횟수
* deathType : 죽거나 살아남은 유형 [ alive, byplayer, byzone, suicide, logout ]
* rideDistance : 총 운전 거리(미터)
* swimDistance : 총 수영 거리(미터)
* walkDistance : 총 걸은 거리(미터)
* timeSurvived : 살아있던 총 시간(초)
* weaponAcquired : 무기를 주운 횟수
* killPlace : 총 킬수 rank (동일한 킬이라면, roster_rank은 낮고, timeSurvived 길수록 killplace 하락) 
* winPlace : rank와 유사하지만, 중간에 나갔다면 나가는 그 순간의 순위(roster_rank는 최종 match 랭크)
